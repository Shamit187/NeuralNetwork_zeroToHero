```python
names = []
with open('names.txt') as file:
    for line in file:
        names.append(line.strip())

print(len(names))
```

```python
import numpy as np

probabilities = np.ones((27,27))
ctoi = lambda c: ord(c.upper()) - ord('A') if c.isalpha() else 26
itoc = lambda i: chr(i + ord('A')) if i < 26 else ' '

for name in names:
    names = '$' + name + '$'
    for lead, trail in zip(names, names[1:]):
        probabilities[ctoi(lead), ctoi(trail)] += 1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(
    probabilities,
    annot=True,
    annot_kws={"size": 5},
    fmt=".0f",
    ax=ax,
    xticklabels=[itoc(i) for i in range(27)],
    yticklabels=[itoc(i) for i in range(27)],
    cbar=False,
    cmap='Blues'
)

# plt.show()
```

```python
# Predicting the next letter
from random import choices
from string import ascii_uppercase


def next_letter(letter):
    i = ctoi(letter)
    distribution = probabilities[i]
    return itoc(choices(range(27), distribution)[0])


def name_generator():
    name = ""
    ch = "$"
    while ch != " ":
        ch = next_letter(ch)
        name += ch

    return name


for _ in range(10):
    print(name_generator())
```
